# MINI PROJECT - 1
Create a console app that allows users to log in and make diary entry. To keep things simple, app only allows user to add an entry or read an earlier entry (skip the edit/delete feature)

Create an authentication system:

Create a sign up system. On starting the app, it should ask user of they want to log in or sign up (By pressing 1 or 2)
If a user decides to sign up , ask for a username and password and save them in a file called passwords.txt
For example:
Passwords.txt :
UserA: password123
UserB: 321drowssap
... 

         Also, while making the entry, make sure the username doesn’t already exist, otherwise show an error saying: “The username already exists”

         Along with this the system should also create a folder with the same username where the user’s diary entries will be saved.


For login flow:
If user chooses to login, ask username and password and verify whether username exists in the file:
Username doesn’t exists: show error “No user found”
Password doesn’t match: show error “Incorrect Password”

If both are correct, show user options to
1) read their entries
2) make an entry to the diary for today


View an entry:
The list will be shown which has file names for all their previous entries (saved with file name as the date on which it was saved)


Make and entry:
The user should be able to make a diary entry.
User can press three “Enter”s in a row to stop making the entry
Once the user is done with entry and presses the exit character/combination, save the entry in the folder you created for that user at sign up time, in a file name same as the current date (whatever format of date suites you)

* For all the flows, whenever a menu is shown to the user, don’t forget to add a ‘back’ option to the menu if applicable.

```python
import os
import datetime

# File to store passwords
password_file = "passwords.txt"

def sign_up():
    # Ask for username and password
    username = input("Enter a username: ")
    password = input("Enter a password: ")

    # Check if username already exists
    with open(password_file, "r") as file:
        for line in file:
            if line.startswith(username):
                print("The username already exists")
                return

    # Add the new user's credentials to the password file
    with open(password_file, "a") as file:
        file.write(f"{username}:{password}\n")

    # Create a folder for the new user's diary entries
    os.mkdir(username)
    print("Sign up successful!")

def log_in():
    # Ask for username and password
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    # Verify the user's credentials
    with open(password_file, "r") as file:
        for line in file:
            if line.startswith(username):
                stored_password = line.strip().split(":")[1]
                if password == stored_password:
                    user_menu(username)
                    return
                else:
                    print("Incorrect password.")
                    return
    print("No user found.")

def user_menu(username):
    while True:
        # Show menu options
        print("Welcome, ", username,"\n" )
        print("1) Read diary entries\n2) Make a diary entry\n3) Back\n")
        option = input("Enter a choice: ")

        if option == "1":
            read_entries(username)
        elif option == "2":
            make_entry(username)
        elif option == "3":
            break
        else:
            print("Invalid option.")

def read_entries(username):
    entries = os.listdir(username)
    # Show all the available entries
    print("Diary Entries\n")
    for entry in entries:
        print(entry)
    input("\nPress Enter to go back...")

def make_entry(username):
    # Ask for the diary entry
    entry = ""
    while True:
        line = input("Enter a line for your diary entry (press Enter three times to finish):\n")
        if line == "":
            if entry == "":
                return
            else:
                break
        entry += line + "\n"
    # Save the entry with the current date as the filename
    today = datetime.date.today()
    filename = today.strftime("%d-%m-%Y")
    with open(f"{username}/{filename}", "w") as file:
        file.write(entry)
    print("Diary entry saved!")

while True:
    # Show main menu options
    print("1) Log in\n2) Sign up\n3) Exit\n")
    option = input("Enter a choice: ")

    if option == "1":
        log_in()
```

This script is a simple implementation of a diary application where users can sign up, log in, and create diary entries. Upon execution, the script presents the user with a main menu with three options: log in, sign up, and exit. If the user chooses to sign up, they are prompted to enter a username and password, which are then stored in a text file. If the user chooses to log in, they are prompted to enter their username and password, which are then verified against the stored credentials. If the user is successfully logged in, they are presented with a user menu where they can read their diary entries, make new diary entries, or log out. The diary entries are stored in individual text files within a folder named after the user.
