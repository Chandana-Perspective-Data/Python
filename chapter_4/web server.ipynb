# web servers

A web server is a piece of software that listens for requests from clients and sends back responses. These requests and responses are typically sent using the Hypertext Transfer Protocol (HTTP), a standard set of rules for exchanging data on the web.

Web servers are important because they host websites and enable them to be accessed over the internet. When you enter a web address into your browser, your computer sends a request to the server that hosts the website you want to visit. The server then sends back the content of the website, which your browser displays on your computer.

There are many different types of web servers, but some of the most commonly used ones include Apache, NGINX, and Microsoft IIS. Each of these servers has its own strengths and is suitable for different types of websites and environments.

In addition to hosting websites, web servers can also provide a variety of other services, such as:

> File transfer: Web servers can be configured to allow users to download or upload files using tools such as FTP (File Transfer Protocol).

> Email: Some web servers include built-in support for sending and receiving email.

> Database access: Web servers can be used to host databases and provide access to them through web applications.

> Streaming media: Web servers can be used to stream audio and video content to users.

Overall, web servers play a crucial role in the modern internet, enabling people to access a wide variety of information and services online.

The World Wide Web is about communication between web clients and web servers.

# HTTP Request / Response
Communication between clients and servers is done by requests and responses:

1.A client (a browser) sends an HTTP request to the web

2.A web server receives the request

3.The server runs an application to process the request

4.The server returns an HTTP response (output) to the browser

5.The client (the browser) receives the response

# The HTTP Request Circle
A typical HTTP request / response circle:

1.The browser requests an HTML page. The server returns an HTML file.

2.The browser requests a style sheet. The server returns a CSS file.

3.The browser requests an JPG image. The server returns a JPG file.

4.The browser requests JavaScript code. The server returns a JS file

5.The browser requests data. The server returns data (in XML or JSON).

# What is webscraping?
Web scraping, web harvesting, or web data extraction is the process of retrieving or "scraping" data from a website. It involves making HTTP requests to a website's server, downloading the HTML of the web page, and parsing that HTML to extract the data you need.

Web scraping is commonly used for a variety of purposes, such as data mining, data analysis, and automated testing. It is a useful tool for extracting information from websites that do not have APIs (Application Programming Interfaces) or when you want to gather data from multiple websites in a structured format.

However, web scraping can be a violation of a website's terms of service, and it may be illegal in some jurisdictions. It is important to be respectful of the website's terms of service and to only scrape data that you have permission to access.
