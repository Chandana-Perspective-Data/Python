```python
import os
from datetime import date



# Function to get the current date in a custom format
def get_date():
    return date.today().strftime("%Y-%m-%d")

# Function to sign up a new user
def sign_up():
    print("Enter a new username and password to sign up:")
    username = input("Username: ")
    password = input("Password: ")
    # Check if the username already exists
    if os.path.exists(f"{username}/"):
        print("The username already exists")
    else:
        # Create a new directory for the user's diary entries
        os.mkdir(username)
        # Save the username and password to a file
        with open("passwords.txt", "a") as f:
            f.write(f"{username}:{password}\n")
        print("You have successfully signed up.")

# Function to log in an existing user
def log_in():
    print("Enter your username and password to log in:")
    username = input("Username: ")
    password = input("Password: ")
    # Check if the username exists and the password is correct
    with open("passwords.txt", "r") as f:
        for line in f:
            if line.startswith(username):
                if line.strip().split(":")[1] == password:
                    print("You have successfully logged in.")
                    return username
                else:
                    print("Incorrect password.")
                    return None
        print("No user found.")
        return None

# Function to read the user's previous entries
def read_entries(username):
    print("List of previous entries:")
    entries = os.listdir(username)
    for entry in entries:
        print(entry)

# Function to make a new diary entry
def make_entry(username):
    entry = ""
    print("Write your entry below:")
    while True:
        line = input()
        if line == "":
            entry += "\n\n"
        else:
            entry += line + "\n"
        if entry.endswith("\n\n\n"):
            break
    # Save the entry to a file
    with open(f"{username}/{get_date()}.txt", "w") as f:
        f.write(entry)
    print("Your entry has been saved.")

# Main program loop
while True:
    print("Welcome to the Diary App.")
    choice = input("Press 1 to log in, 2 to sign up, or 3 to quit: ")
    if choice == "1":
        username = log_in()
        if username is not None:
            while True:
                print(f"Welcome, {username}.")
                choice = input("Press 1 to read your entries, 2 to make a new entry, or 3 to log out: ")
                if choice == "1":
                    read_entries(username)
                elif choice == "2":
                    make_entry(username)
                elif choice == "3":
                    break
                else:
                    print("Invalid choice.")
    elif choice == "2":
        sign_up()
    elif choice == "3":
        break
    else:
        print("Invalid choice.")
```

Here is a list of some basic rules (except splitting and doubling):

The goal of blackjack is to beat the dealer's hand without going over 21.

Face cards are worth 10. Aces are worth 1 or 11, whichever makes a better hand.

Each player starts with two cards, one of the dealer's cards is hidden until the end.

To 'Hit' is to ask for another card. To 'Stand' is to hold your total and end your turn.

If you go over 21 you bust, and the dealer wins regardless of the dealer's hand.

If you are dealt 21 from the start (Ace & 10), you got a blackjack (instant win).

Blackjack usually means you win 1.5 the amount of your bet. Depends on the casino.

Dealer will hit until his/her cards total 17 or higher i.e once the player chooses to hold (and they aren't busted yet) the dealer has to open the hidden card and keep hitting until the sum is 17 or more

If dealer gets busted, player wins


It is advised to play some games online/ read more about it before implementing.
