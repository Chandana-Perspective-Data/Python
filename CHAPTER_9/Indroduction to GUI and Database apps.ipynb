# Indroduction to GUI and Database apps

A GUI (Graphical User Interface) is a type of user interface that allows users to interact with electronic devices through graphical icons and visual indicators, as opposed to text-based interfaces. A GUI provides a way for users to interact with software and applications through visual representations of commands and options, such as buttons and icons.

A database app is a software application that interacts with a database to store, retrieve, and manage data. Database apps can range from simple desktop apps for personal use to complex enterprise-level applications that manage vast amounts of data and serve many users simultaneously. Some common examples of database applications include customer relationship management (CRM) systems, inventory management systems, and financial reporting systems.

By combining a GUI with a database app, users are able to interact with a database in a visual, intuitive manner, making it easier to manage and manipulate data.





# USING GUI TKINTER /JINJA


Tkinter is a GUI library for Python and it is the standard GUI library for Python. It provides a simple way to create windows, dialogs, buttons, and other graphical user interface elements. Tkinter is easy to use and comes bundled with the Python distribution, so there is no need to install any additional packages.

Jinja is a template engine for Python that is used for generating HTML, XML, or other markup. Unlike Tkinter, Jinja is not a GUI library, but it can be used in combination with a GUI library like Tkinter to create dynamic, web-based applications. Jinja provides a simple syntax for defining templates and allows developers to include variables and dynamic content in their templates, making it easy to create dynamic web pages that can be updated based on user input or other data.

So, Tkinter provides the GUI capabilities, while Jinja is used for creating dynamic templates that can be displayed in the GUI. Together, Tkinter and Jinja can be used to create dynamic, web-based applications with graphical user interfaces.

# Tkinter provides the following geomerty methods
1) pack() method is used to organize componenets or wedgets in main window.

synatx:
    
    widget.pack(options)
The possible options are :

side: it represents the side to which the widget is to be placed on the window.Side may be LEFT or RIGHT or TOP(default) or BOTTOM.



```python
from tkinter import *
top =Tk()
top.geometry("300x200")
btn1=Button(top,text="login")
btn1.pack(side=LEFT)
top.mainloop()
```

2) grid() method:

the grid() method organize the widgets in the tabular form. we can specify the rows and columns as the options in the method call.

this is a more organize way to place the widgets to the python applications.

synatx:
  widget.grid(options)
  
  
  
The possible options are

.Column 

 the column number in which the widget is to be placed . The leftmost column is represented by 0.
 
. padx,pady
 
 it represented the number of pixels to pad the widget outside the wedget's border.
 
. row
 
 the row number in which the widget is to be placed . The topmost row is represented by 0.
 


```python
# example
from tkinter import*
parent = Tk()
parent.title("student")
parent.geometry("400x300")
name=Label(parent,text="Name:")
name.grid(row=0,column=0,padx=10,pady=5)
e1=Entry(parent)
e1.grid(row=0,column=1)
regno=Label(parent,text="Regd No :")
regno.grid(row=1,column=0,padx=10,pady=5)
e2=Entry(parent)
e2.grid(row=1,column=1)
btn=Button(parent,text="submit")
btn.grid(row=3,column=1)
parent.mainloop()
```

3) place() method :

the place() method organizes the widgets to the specific x and y coordinates:
 
 
synatx:
  
  widget.place(x,y)
 
. x,y :
  
  it refers to the horiontal and vertical offset  in the pixels

```python
from tkinter import *
parent= Tk()
parent.title("student")
parent.geometry("300x200")
name=Label(parent,text="Name :")
name.place(x=50,y=50)
e1=Entry(parent)
e1.place(x=100,y=50)
regno =Label(parent,text="Regd No:")
regno.place(x=50,y=100)
e2=Entry(parent)
e2.place(x=100,y=100)
parent.mainloop()
```

# Button Widget in Tkinter:
. The Button is used to add various kinds of buttons to the python applications.We can also associate a method or function with 
a button which is called when the  button is pressed.

synatx: 
     
        name=Button(parent,option)

the options are

. activebackground: it represents the background of the button when it is active.
    
. activeforeground: it represents the fore color of the button when it is active.

. bd: it represents the border width in pixels
    
. bg: it represents the background color of the button.
    
. command: it is set to the function call which is scheduled when the function is called.
    
.text: it is set to the text displayed on the button.
    
.fg: foreground color of the button
    
.height: the height os the button
    
.padx: Additional padding to the button in the horizontal direction
    
.pady: Additional padding to the button in the vertical direction
    
.width: the width of the button.

```python
from tkinter import*
from tkinter import messagebox
top=Tk()
top.geometry("300x200")

def fun():
    messagebox.showinfo("hello","blue Button clicked")
btn1=Button(top,text="Red",bg="red",fg="white",width=10)
btn1.pack(side=LEFT)
btn2=Button(top,text="Green",bg="green",fg="white",width=10,height=5,activebackground ="yellow")
btn2.pack(side=TOP)
btn3=Button(top,text="Blue",bg="blue",fg="white",padx=10,pady=10,command=fun)
btn3.pack(side=BOTTOM)
top.mainloop()
```

# Entry widget in Tkinter:
 
 
. the entry widget is used to provide the single line text-box to the user to accept a value from the user. we can use the entry
 widget to accept the text strings from the user.
 
 syntax:
  
  name = Entry(parent,options)
  
The options are:

. bd: it represents the border width in pixels

.bg: it represents the background color of the Entry.

.show :  it is used to show the entry text of some other type instead of the string . for example,
 
 the password is typed using stars(*)
 
. fg: foreground color of the entry.

.width:  the width of the entry.


```python
from tkinter import*

top=Tk()
top.geometry("300x200")
   
enty0=Entry(top,width="30")
enty0.place(x=50,y=40)

enty1=Entry(top,bg="Yellow")
enty1.place(x=50,y=70)

enty2=Entry(top,fg="red",show="*")
enty2.place(x=50,y=100)
top.mainloop()
```

# Frame Widget in Tkinter :
. Frame widget is used  to organize the group of widgets.it acts like a container which can be used to hold the other widgets.
the rectangular areas of the screen are used to organize the widgets  to the python application.

syntax:
    
    name=Frame(parent,options)
    
the options are:
    
.bd: it represents the border width in pixels.

.bg: it represents the background color of the frame.
    
.width:   the width of the frame.
    
.height:  the height of the frame

```python
from tkinter import*
top=Tk()
top.geometry("300x200")
tframe=Frame(top,width="100",height="100",bg="yellow")
tframe.pack()

lframe=Frame(top,width="100",height="50",bg="blue")
lframe.pack(side=LEFT)

rframe=Frame(top,width="100",height="50",bg="green")
rframe.pack(side=RIGHT)

btn1=Button(tframe,text="Submit",fg="red")
btn1.place(x=10,y=10)
top.mainloop()
```
